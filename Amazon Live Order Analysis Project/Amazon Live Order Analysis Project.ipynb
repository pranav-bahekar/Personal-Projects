```python
import pandas as pd
data = pd.read_csv('amazon_order_history2.csv')
data.head()
```

```python
data['total']=data['total'].str.replace('Rs.','')
```

```python
data
```

```python
data['total']=data['total'].str.replace(',','').astype('float')
```

```python
data
```

```python
data['total'].sum()   #THIS MUCH I SPEND IN THE YEAR 2021
```

```python
data['total'].mean()    #Average ON HOW MUCH I SPEND in 2021
```

```python
data['total'].min()   #Minimum Spent
```

```python
data['total'].max()   #Maximum Spent
```

```python
data[data['total']==25990.0][['items','total']]  #Item which was having max price
```

```python
data['date']=pd.to_datetime(data['date'])
```

```python
data.head()
```

```python
data.plot.bar(x='date',y='total',figsize=(10,8))
```

```python
do=data.groupby('date').sum()['total']
do
```
